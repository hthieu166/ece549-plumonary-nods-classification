```python
import pandas as pd
import os as os
import numpy as np
```

```python
path = 'F:\\医学数据集\\LUNA\\rowfile\\subset5'
LUNA_data = pd.read_csv('../data/annotationdetclsconvfnl_v3.csv').to_numpy()
srsid = LUNA_data[:,0]
test_mhd_name = os.listdir(path)
test_mhd_name_prefix = list(map(lambda x: x[:-4], test_mhd_name))
def is_test_data_file(data):
    return data.split('-')[0] in test_mhd_name_prefix
def is_train_data_file(data):
    return data.split('-')[0] not in test_mhd_name_prefix
def is_test_data(data):
    return data[0].split('-')[0] in test_mhd_name_prefix
def is_train_data(data):
    return data[0].split('-')[0] not in test_mhd_name_prefix
print(srsid.shape)
```

```python
test_data_file_name = list(filter(is_test_data_file, srsid))
train_data_file_name = list(filter(is_train_data_file, srsid))
test_data = list(filter(is_test_data,LUNA_data))
train_data = list(filter(is_train_data,LUNA_data))
```

```python
print(test_data_file_name)
print(len(test_data_file_name))
print(train_data_file_name)
print(len(train_data_file_name))
```

```python
print(len(test_data))
print(len(train_data))


```

```python
test_data = np.array(test_data)
train_data = np.array(train_data)
```

```python
def print_status(data):
    print(f'data count :{data.shape[0]}\n'
          f'positive data count:{np.sum(data[:,-1])}\n'
          f'negative data count:{data.shape[0]-np.sum(data[:,-1])}')
```

```python
print('test')
print_status(test_data)
```

```python
print('train')
print_status(train_data)
```
