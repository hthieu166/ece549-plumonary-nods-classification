# Create df based on .pkl file, features selected, and identifiers for luna

```python
f = pyfea_path + '/LIDC-IDRI-0001/1/IL057_127364.pkl'
with open(f, 'rb') as f:
    data = pickle.load(f)    
ex_list = list(np.array(list(data.values())[22:]))
print(len(ex_list), ex_list)
```

```python
def find_best(a):
    mean_a = np.mean(a, axis=0)
    err = []
    for row in a:
        err.append(np.sum((row-mean_a)**2))
    return list(a[np.argmin(err)]), np.argmin(err)

a = [[23.19226498, 0.760052393,1695.118914,2.80418334,298.7252343,-0.778248523,5.468781189,1606.136914,40.60655738,0.185816602],
    [15.92659051,0.716648904,1696.972853,2.287992269,335.0043031,-0.625141891,5.599244265,1195.814177,38.95833333,0.255425157],
    [23.26549303,0.761653668,1697.590833,2.801704218,298.7512154,-0.777255596,5.468588132,1605.555768,40.6416882,0.185816618]]
a = np.array(a)
print(np.shape(a))
print(list(np.mean(a, axis=0)))
print(list(np.median(a, axis=0)))
print(find_best(a))
```

```python
lidc_id = []
roi_id = []
pkl_id = []
features = []

new_lidc = ['LIDC-IDRI-0101', 'LIDC-IDRI-0123', 'LIDC-IDRI-0660','LIDC-IDRI-0820']

lidc_folders = glob(pyfea_path+'/*')
m=0
# loop through all LIDC-IDRI folders
for lidc_folder in lidc_folders:
    rois = glob(lidc_folder+'/*')
    
    # loop though all roi folders
    for roi in rois:
        pkl_files = glob(roi+'/*')
        lidc_id.append(lidc_folder.split('/')[-1])
        roi_id.append(roi.split('/')[-1])
        
        # loop through all pkl files
        feature_tmp = []
        pkl_tmp = []
        for pkl in pkl_files: 
            # append identifiers for df
            pkl_tmp.append(pkl.split('/')[-1])
            
            # open pkl files and load data values
            with open(pkl, 'rb') as f:
                data = pickle.load(f)  
            
            if (lidc_id[-1] in new_lidc):
                feature_tmp.append(list(np.array(list(data.values())[32:])))
            else:
                feature_tmp.append(list(np.array(list(data.values())[22:])))
            
        # if there are pkl files
        if (len(feature_tmp)>0):
            # mean mode
            if (mode == 'mean'):
                pkl_id.append('mean')
                features.append(list(np.mean(feature_tmp, axis=0)))
            # median mode
            if (mode == 'median'):
                pkl_id.append('median')
                features.append(list(np.median(feature_tmp, axis=0)))
                
            # pick features based on l2 error from mean
            if (mode == 'l2'):
#                 print(lidc_id[-1],roi_id[-1], np.shape(feature_tmp))
                l_feature, idx = find_best(feature_tmp)
                pkl_id.append(pkl_tmp[idx])
                features.append(np.array(l_feature))
        else:
            # print lidc id and roi id for features without pkl files
            print(m, lidc_id[-1],roi_id[-1])
            m+=1
# example:
# print(lidc_id[:10])
# print(roi_id[:10])
# print(pkl_id[:10])
# print(features[:10])
```
