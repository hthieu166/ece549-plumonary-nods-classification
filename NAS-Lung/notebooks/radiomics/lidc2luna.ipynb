## Combine luna annotation and LIDC using SeriesInstanceUID

```python
lidc2luna_df = pd.read_csv('mappingLUNA16.csv', delim_whitespace=True)
lidc2luna_df
```

```python
SeriesInstanceUID = []
for index, row in anno_df.iterrows():
    SeriesInstanceUID.append(row['seriesuid'].split('-')[0])
anno_df['SeriesInstanceUID']=SeriesInstanceUID
anno_df
```

```python
comb_df = pd.merge(anno_df, lidc2luna_df, on=["SeriesInstanceUID", "SeriesInstanceUID"])
comb_df
```

```python
list32_df = pd.read_csv("list32.csv")
list(list32_df)
```

## Match by diameter from list32.csv and csv just obtained from above

```python
dia_dict = {}
for index, row in list32_df.iterrows():
    case = 'LIDC-IDRI-' + ('%04d' % int(row['case']))
    if (case not in dia_dict.keys()):
        dia_dict[case] = {}
    dia_dict[case][row['eq. diam.']] = row['roi']
dia_dict
```

```python
approx_roi = []
case_list = []

error_dia = []
list32_dia = []
error_dia2 = []
list32_dia2 = []

def best_dia(list_dia, dia):
    list_dia = np.asarray(list_dia)
    idx = (np.abs(list_dia - dia)).argmin()
    return list_dia[idx], np.abs(list_dia[idx] - dia), idx

for index, row in comb_df.iterrows():
    case = row['PatientID']
    list_dia = list(dia_dict[case].keys())
    dia, err, i = best_dia(list_dia, row['diameter_mm'])

    dia2 = -1
    err2 = -1
    if (len(list_dia)>1):
        list_dia.remove(dia)
        dia2, err2,_ = best_dia(list_dia, row['diameter_mm'])
    
    approx_roi.append(int(dia_dict[case][dia]))
    case_list.append(case.split('-')[-1])  

    error_dia.append(err)
    list32_dia.append(dia)
    error_dia2.append(err2)
    list32_dia2.append(dia2)

comb_df['approx_roi'] = approx_roi
comb_df['case'] = case_list
comb_df['error_dia'] = error_dia
comb_df['list32_dia'] = list32_dia
comb_df['error_dia2'] = error_dia2
comb_df['list32_dia2'] = list32_dia2

comb_df
```

```python
comb_df.to_csv('lidc2luna.csv')
```
